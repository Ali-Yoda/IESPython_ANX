```python
class Pratchett:
    '''
    Downloader manager for Pratchett characters
    
    It contains methods for collection of links, downloading itself and storing results
    '''
    def __init__(self, link):
        '''
        Creates the Pratchett object with self.links that store links to be downloaded
        to store books and characters of the Pratchett
        '''
        self.link = link
        
    def getSoup(self):

        r = requests.get(self.link)
        r.encoding = "UTF-8"
        soup = BeautifulSoup(r.text, "lxml")
        return soup
    
    def getAllLinks(self):
        '''Downloads all characters links.'''
        soup2 = self.getSoup().find('span',{'id':'Unseen_University_and_the_Wizards'}).parent.previous.previous.parent.parent.parent.find_all('li')
        links2 = ["http://wiki.lspace.org" + s.find("a")["href"] for s in soup2]
        return links2
    
    def getName(self):
        '''Extract the name of the character.'''
        r2 = []
        for li in self.getAllLinks():
            r3 = requests.get(li)
            r3.encoding = "UTF-8"
            r4 = BeautifulSoup(r3.text, "lxml").find("title").text.split("-")[0].strip()
            r2.append(r4)
            
        return r2
    
    def getBooks(self):
        # not functioning - including cameos
        r5 = []
        for li in self.getAllLinks():
            r6 = requests.get(li)
            r6.encoding = "UTF-8"
            r7 = BeautifulSoup(r6.text, "lxml").findAll("i")
            r8 = [s.text.strip() for s in r7]
            r5.append(r8)
        return r5
    
    def saveDFs(self):
        '''Generate Panda Data Frame.'''
        names = self.getName()
        df = pd.DataFrame({
            "character_name": names })
        
        return df

        
```
